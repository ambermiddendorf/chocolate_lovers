```python
import pandas as pd
from path import Path
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()

file_path = Path("chocolate.csv")
chocolate_df = pd.read_csv(file_path)
chocolate_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = chocolate_df.copy()
df2['cocoa_butter'] = le.fit_transform(df2['cocoa_butter'])
df2['vanilla'] = le.fit_transform(df2['vanilla'])
df2['lecithin'] = le.fit_transform(df2['lecithin'])
df2['salt'] = le.fit_transform(df2['salt'])
df2['sugar'] = le.fit_transform(df2['sugar'])
df2['sweetener_without_sugar'] = le.fit_transform(df2['sweetener_without_sugar'])
df2.head(50)
```
