```python
# make df for all words with Index
txt_df = pd.DataFrame(columns=['Index', 'Taste'])  
```

```python
first_df = df.loc[df.first_taste.isnull()==False, ['Index','first_taste']]
```

```python
sec_df = df.loc[df.second_taste.isnull()==False, ['Index','second_taste']]
```

```python
third_df = df.loc[df.third_taste.isnull()==False, ['Index','third_taste']]
```

```python
forth_df = df.loc[df.fourth_taste.isnull()==False, ['Index','fourth_taste']]
```

```python
txt_df = pd.concat([first_df.rename(columns={'first_taste':'Taste'}), sec_df.rename(columns={'second_taste':'Taste'}), third_df.rename(columns={'third_taste':'Taste'}), forth_df.rename(columns={'fourth_taste':'Taste'})], ignore_index=True)
```

```python
txt_df
```

```python
txt_df.columns
```

```python
words_df = pd.DataFrame()
i=0
for index, row in txt_df.iterrows():
    words = row.Taste.split(" ")
    ind = row.Index
    for word in words:
        df2=pd.DataFrame({'Index': ind,'Word': word}, index=[i])        
        words_df = pd.concat([words_df, df2], ignore_index=True, axis =0)
#        words_df.append({'Index': ind,'Word': word},ignore_index = True)
#        words_df.append(row_dict, ignore_index=True)
#    break
#        words_df.append({'Index': ind,'Word': word}, ignore_index = True)
```

```python
words_df
```

```python
words_df.Word = words_df.Word.str.lower()
```

```python
# initialize spell check model
ob = spell_checker(words_df, 'Word')
```

```python
# quick analysis
ob.spell_check()
```

```python
error_df = pd.DataFrame(ob.get_all_errors())
```

```python
error_df.rename(columns={0: 'errors'}, inplace=True)
```

```python
error_grps = pd.DataFrame(error_df.groupby('errors').size(), columns=['Count'])
```

```python
error_grps.sort_values('Count', ascending=False)
```

```python
error_grps.to_csv("error_grps.csv")
```

```python
corrections = pd.read_csv("error_corrects.csv")
corrections.rename(columns={'errors':'Word'}, inplace = True)
```

```python
corrections
```

```python
merged = pd.merge(words_df, corrections, on="Word", how="left")
```

```python
# fixed_df = merged[['Index', Word = lambda x: (x.Corrected if x.Corrected != 'NaN' else x.Word, axis=1)]]
merged['final'] = np.where(merged['Corrected'].isnull(), merged['Word'], merged['Corrected'])
```

```python
ratings = df[['Index', 'rating']]
```

```python
taste_ratings = pd.merge(words_df, ratings, how="left")
taste_ratings
```

```python
def get_rating_group(taste_ratings):
    if taste_ratings['rating'] >= 4:
        return 4
    elif taste_ratings['rating'] >= 3:
        return 3
    elif taste_ratings['rating'] >= 2:
        return 2
    elif taste_ratings['rating'] >= 1:
        return 1
    else:
        return 0
        
taste_ratings['ratings_group']= taste_ratings['rating'].apply(lambda x: round(x,0))
```
